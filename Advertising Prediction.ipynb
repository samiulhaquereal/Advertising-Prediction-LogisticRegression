```python
df=df.drop(['Timestamp'], axis = 1)
```

```python
df
```

```python
x = df.drop(['Clicked on Ad'], axis = 1)
```

```python
x
```

```python
y = df['Clicked on Ad']
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le_x= LabelEncoder()
x.City = le_x.fit_transform(x.City)
x.Country = le_x.fit_transform(x.Country)
```

```python
x
```

```python
x.rename(columns={'Ad Topic Line': 'Ad_Topic_Line'}, inplace=True)
```

```python
x.Ad_Topic_Line = le_x.fit_transform(x.Ad_Topic_Line)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression()
reg.fit(xtrain,ytrain)
```

```python
prediction = reg.predict(xtest)
```

```python
prediction
```

```python
reg.score(xtest,ytest)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,prediction))
```

```python
from sklearn.metrics import  confusion_matrix
```

```python
print(confusion_matrix(ytest, prediction))
```
